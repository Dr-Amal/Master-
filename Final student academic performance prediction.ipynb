```python
def Oversampling(data,samples):
        import pandas as pd
        data=data
        y=samples
        count_class_0, count_class_2, count_class_1 = samples.value_counts()
        data_class_2 = data[data[y.name] == 2]
        data_class_0 = data[data[y.name] == 0]
        sample1=data.loc[data['StageID'] == 1]
        print ('Class 2 = ',data_class_2.shape,'Class 0 =',data_class_0.shape)
        data_class_2_over = data_class_2.sample(count_class_0, replace=True)
        data_test_over = pd.concat([data_class_0,data_class_2_over,sample1], axis=0)
        print('Random over-sampling:')
        data_test_over.to_excel('newdataset.xlsx',index=False)
        data=pd.read_excel('newdataset.xlsx')
        return (data[y.name].value_counts()),data.shape
```
